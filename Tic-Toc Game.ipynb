```python
import random

# List of Board from 1,10 upto 10 not 10 starting from 1 to ignore Zero position

#Global Variable
board=[' ']*10



# Display Tic Tac board using the list board
def display_board(board_psn):
    '''
    to Print the Tic-Tac Board as per the positions in out list called board
    '''
    #clear_output()
    print(' '+board_psn[7]+' | '+board_psn[8]+' | '+board_psn[9])
    print("----------")
    print(' '+board_psn[4]+' | '+board_psn[5]+' | '+board_psn[6])
    print("----------")
    print(' '+board_psn[1]+' | '+board_psn[2]+' | '+board_psn[3])
    
# Randomly ask player his/her choice of X or O -- function =choose_plyer
def choose_plyer():
    '''
    Randomly ask the player to choose Marker 'X' or 'O'
    random.randint function is used
    return Player1 or Player2
    '''
    ans=random.randint(0,1)
    if ans ==0:
        return 'Player1'
    else:
        return 'Player2'

# While loop till player choose X or O and return the tuple with Player1 and Player2
def player_sel():
    '''
        function player_sel
        Player need to choose his/her Sigh (X or O)
        Function retuen tuple with ('player1's sign','player2's sign') 
    '''    
    marker=''
    while marker.upper() !='X' and marker.upper() != 'O':
        marker=input(f" Player 1 Choose X or O = ")
    if marker.upper()=='X':
        return('X','O')
    else:
        return('O','X')


#To Update the postion of X and Y on board as per player's choice
def place_mrkr(board,marker,position):
    '''
    -As per Players mark i.e sign 'X' or 'Y' list board's position will get updated 
    -Board is our board list,Marker is X or O and position is selection of player in between 1 to 9
    '''
    board[position]=marker
# To check the win condition of both the Players
def win_check(board,mark):
    '''
    TO check WIN condition of game
    -All ROWS and check to see if they all share the same mark?
    -All COLUMNS,CHeck to see if all mark matches
    -2 Diagonals,Check to match the mark
    '''
    return (
    #All ROWS and check to see if they all share the same mark?
    (board[1] == mark and board[2] == mark and board[3] == mark) or
    (board[4] == mark and board[5] == mark and board[6] == mark) or
    (board[7] == mark and board[8] == mark and board[9] == mark) or
    #All COLUMNS,CHeck to see if all mark matches
    (board[1] == mark and board[4] == mark and board[7] == mark) or
    (board[2] == mark and board[5] == mark and board[8] == mark) or
    (board[3] == mark and board[6] == mark and board[9] == mark) or
    #2 Diagonals,Check to match the mark
    (board[1] == mark and board[5] == mark and board[9] == mark) or
    (board[7] == mark and board[5] == mark and board[3] == mark)
    )

#Comapre the Player's selection with the Board list to check the position is available/not
def space_check(board,position):
    '''
    Return boolean value if space is availabe on the board
    Basically check the (positions) options available for both the players
    '''
    return board[position] == ' '



#Check the board list 1 to 10 and use space_check function to check the positions are available/not
def full_board_check(board): 
    '''
    Returns True If both players don't have any space/position to select on board
    will check the board using for loop and call space_check function to check space
    '''
    for i in range(1,10): #uptp 9 not including 10 start from 1 ignore '#'
        if space_check(board,i):  #False means We have space and board is not full
            #if space_check is false means we have space/postions
            return False #False- Board is not full
    return True #if condition is True means we dont have space [0,10]

#function ask for players next position and uses space _check function
#If choise is not in 1-9 and space_check is not null then it will ask you for your 
#  options again and again (while loop)
def player_posotion(board,turn):
    '''
    Ask the player its position. 
    this function will validate choice should be in 1-9 and spacecheck is not space ' '.
    '''
    position = 0
    while position not in range(1,10) or not space_check(board,position):
        position=int(input(f'{turn} Choose a position (1-9) :- '))
    return position
#What to replay the game 'Yes' or 'No' return boolean function
def reply():
    choise = input("Play Again? Enter 'Y' for Yes \n")
    return choise.upper()=='Y'
        
#While loop to keep running  the game

print("------------Tic-Toc Game------------") 
print("No of choices both the players have with the input options (1 to 9)")

while True:
    #Play the same
    ##set who is first,Board,choose marker of respective player(X,O)
    
    ##Game Play
    board=[' ']*10
    # Assign X or O to Player1 and Player2 slection function  player_sel
    player1_marker,player2_marker=player_sel()
    print(f"Player1={player1_marker}   and Player2={player2_marker}")
    
    # choose_plyer function to choose whos going first
    turn=choose_plyer()
    print(' '+ turn +" will go first..!")
    
    game_on=(input("Ready to play? Y or N \n"))
    if game_on.upper() =='Y':
        game_on = True
    else:
        game_on = False
        break
    
    while game_on:
        ##Player 1 Turn
        if turn == 'Player1':
            #display board
            display_board(board)
            #Choose the position 
            position=player_posotion(board,turn)
            #Place the maker to that position
            place_mrkr(board,player1_marker,position)
            #check the win condition
            if win_check(board,player1_marker):
                display_board(board)
                print("Congratulation Player 1 has Won...!!! ")
                #If Player is win then game_on while loop will over
                game_on= False
            #if player 1 is not won then 
            else:
                #check if there is a tie 
                if full_board_check(board):
                    display_board(board)
                    print(" Game is Tie..!!")
                    ##If Tis is there then game_on while loop will over
                    game_on= False
                else:
                    #Game is not tie or no win and board is not full then next players turn
                    turn= 'Player2'
        else:
            ##Player 2 Turn
            if turn == 'Player2':
                #display board
                display_board(board)
                #Choose the position 
                position=player_posotion(board,turn)
                #Place the maker to that position
                place_mrkr(board,player2_marker,position)
                #check the win condition
                if win_check(board,player2_marker):
                    display_board(board)
                    print("Congratulation Player 2 has Won...!!! ")
                    #If Player is win then game_on while loop will over
                    game_on= False
                #if player 2 is not won then 
                else:
                    #check if there is a tie 
                    if full_board_check(board):
                        display_board(board)
                        print(" Game is Tie..!!")
                        ##If Tis is there then game_on while loop will over
                        game_on= False
                    else:
                        #Game is not tie or no win and board is not full then next players turn
                        turn= 'Player1'
    #Break of of while loop on replay()
    if not reply():
        break
```
