```python
import numpy as np
import pandas as pd
from pandas import DataFrame,Series

movies_sub=pd.read_csv('C:\\Users\\735184\\Downloads\\P4-Movie-Ratings.csv')
movies_all=pd.read_csv('C:\\Users\\735184\\Downloads\\dataset.csv',encoding='latin1')

movies_sub.columns=['Movie', 'Genre', 'Ratings', 'AudienceRatings_10','Budget_Million', 'Year']
movies_sub.AudienceRatings_10=movies_sub.AudienceRatings_10/10

# Both conditions returns  the same result
#movies_sub.AudienceRatings_10=np.where(movies_sub.AudienceRatings_10>50,movies_sub.AudienceRatings_10/10,movies_sub.AudienceRatings_10)
#movies_sub.AudienceRatings_10=movies_sub.AudienceRatings_10.mask(movies_sub.AudienceRatings_10 > 50,movies_sub.AudienceRatings_10/10)

movies_all.columns=['DayofWeek', 'Director', 'Genre', 'Movie', 'ReleaseDate','Studio', 'AdjustedGross', 'Budget', 'Gross',\
       'IMDbRating_10', 'MovieLensRating_5', 'Overseas', 'Overseas','Profit','ProfitUS', 'Runtime_min', 'US','GrossUS']

#Join two data frames

merged_data=pd.merge(left=movies_all[['Genre','Movie','ProfitUS','IMDbRating_10', 'MovieLensRating_5']], \
                     right=movies_sub[['Movie','AudienceRatings_10','Budget_Million','Year']],how='inner',on = 'Movie')

#Return Seleceted columns only from Joined data
DataFrame(merged_data,columns=['Genre','Movie','IMDbRating_10','AudienceRatings_10','Year'])

##Create .CSV file with Seleceted columns only from Joined datafarame   TO_CSV function
DataFrame(merged_data,columns=['Genre','Movie','IMDbRating_10','AudienceRatings_10','Year']).to_csv('TESTINGTESTING.csv',sep='|')


##Create table from list
#city=['a','b','c']
#pop=[100,200,30]
#data=DataFrame({'city':np.array(city),'population':np.array(pop)})
#data

##Create table from Dictionary
#data2={'city':['a','b','c'],'pop':[10330,20220,32220]}
#data1=DataFrame(data2)
#data1

```
