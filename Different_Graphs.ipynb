```python
import numpy as np
import pandas as pd
import os
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings ('ignore')

%matplotlib inline
sns.set_style('darkgrid')
#sns.set_style('white')
plt.rcParams['figure.figsize']=10,6


movies=pd.read_csv('C:\\Users\\735184\\Downloads\\P4-Movie-Ratings.csv')

movies.columns=['Film','Genre','Critrc_Ratings_prctg', 'Aud_Ratings_prctg','Budget_mln_DLR', 'Year_Release']
movies.Film=movies.Film.astype('category')
movies.Genre=movies.Genre.astype('category')
movies.Year_Release=movies.Year_Release.astype('category')

#jn =sns.jointplot(data=movies,x='Aud_Ratings_prctg',y='Critrc_Ratings_prctg',kind='hex')
#m1=sns.distplot(movies.Aud_Ratings_prctg,bins=20)
#m2=sns.distplot(movies.Critrc_Ratings_prctg,bins=20)
#n1=plt.hist(movies.Aud_Ratings_prctg,bins=15)
#n2=plt.hist(movies.Critrc_Ratings_prctg,bins=15)
#v4=sns.lmplot(data=movies,x='Critrc_Ratings_prctg',y='Aud_Ratings_prctg',fit_reg=False,hue='Genre',size=7,aspect=1)



hist_lst=[]
label_lst=[]
for mov_typ in movies.Genre.unique():
    label_lst.append(mov_typ)
    hist_lst.append(movies[movies.Genre== mov_typ].Budget_mln_DLR)
    #plt.hist(movies[movies.Genre== mov_typ].Budget_mln_DLR,bins=15,stacked=True) for every Movie Type bins=15

plt.hist(hist_lst,bins=30,stacked=True,label=label_lst)  #for all movie type bins=15 
plt.legend(loc='upper center', bbox_to_anchor=(1.1,1))
plt.show()

# Result for plt.hist(movies[movies.Genre== mov_typ].Budget_mln_DLR,bins=15,stacked=True) "Every movie type bins=15 defined"
# will be different as compaired to
# plt.hist(lst,bins=15,stacked=True) here "for all the list we have defined bins=15" 
sns.set_style('dark')
v5=sns.kdeplot(movies.Critrc_Ratings_prctg,movies.Aud_Ratings_prctg,shade=True,shade_lowest=False,cmap='Reds')
v6=sns.kdeplot(movies.Critrc_Ratings_prctg,movies.Aud_Ratings_prctg,cmap='Reds')

#Here We are overlapping two KDE for Smooth Shades V6 adding Edges to our V5 which is more professional



```
